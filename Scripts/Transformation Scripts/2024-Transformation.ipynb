```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('/Users/badrinathsanagavaram/Desktop/Election Data Analysis/Data/raw-data/2024-election-dataset-raw.csv')
df.head(10)
```

```python
df.drop(columns='Unnamed: 0', inplace=True)
```

```python
df['Type'].replace({
    'OC':'GEN'
},inplace=True)
```

```python
df.head(10)
df.drop(columns='No.', inplace=True)
```

```python
df.head(5)
```

```python
df.rename(columns={
    'Candidate': 'Winning Candidate',
    'Party' : 'Winning Party',
    '%' : 'Winning Percentage',
    'Votes' : 'Winner\'s Votes',
    'Candidate.1': 'Runner Up Candidate',
    'Party.1': 'Runner Up Party',
    '%.1': 'Runner Up Percentage',
    'Votes.1': 'Runner Up Votes',
    '%.2' : 'Margin %',
    'Votes.2': 'Margin Votes'
}, inplace=True)
df.head(5)
```

```python
df.nunique()
```

```python
df.dtypes
```

```python
df['Margin %'] = df['Margin %'].str.replace('%', '').astype(float).astype(int)  
```

```python
df.head(5)
```

```python
df['Runner Up Percentage'] = df['Runner Up Percentage'].str.replace('%', '').astype(float).astype(int)
df['Winning Percentage'] = df['Winning Percentage'].str.replace('%', '').astype(float).astype(int)
```

```python
df.head(5)
```

```python
df['Winner\'s Votes'] = df['Winner\'s Votes'].str.replace(',', '').astype(float).astype(int)
```

```python
df['Margin Votes'] = df['Margin Votes'].str.replace(',', '').astype(float).astype(int)
```

```python
df['Runner Up Votes'] = df['Runner Up Votes'].str.replace('[, ]', '', regex=True).astype(int)
```
