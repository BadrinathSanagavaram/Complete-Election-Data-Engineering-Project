### Replacing the Type Keywords with univeral Type values like in 2024 data transformation dataset and 2019 data transformation dataset.

```python
# first find out the distinct 'Type' values present in the column 
Type_dict = {}
a = dfs[1980]['Type'].unique()
```

```python
year= [1980,1984,1989,1991,1996,1998,1999,2004,2009,2014]
for k in year: 
    a = dfs[k]['Type'].unique()
    type_set = []
    for i in a :
        type_set.append(i)
    Type_dict[k] = type_set
Type_dict
```

```python
## Now in the year 1991 we have another Type which is 'N.A' and we need to find what exactly it is and is there any to replace it

filt = dfs[1991]['Type'] == 'N.A'
dfs[1991].loc[filt]
```

```python
## in the year 1991 in the lok sabha constituncy of Meerut the election was postponed due to Political violence so we drop that row

index = dfs[1991].loc[filt].index
dfs[1991].drop(index=index, inplace=True)
```

### Converting the Winners Votes and Runner's Votes into int variable type

```python
# Before doing that we are checking whether they are int already or not

var_dict = {}
for k in year:
    var_col_dict = {}
    var_list = []
    for col in dfs[k].columns:
        var_col_dict[col] = dfs[k][col].dtypes
    var_dict[k] = var_col_dict
var_dict
```
