```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.elections.in/parliamentary-constituencies/2009-election-results.html"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
all_tables = soup.find_all('table')
len(all_tables)
```

```python
main_table = all_tables[3]
main_table
```

```python
header = [i.text.strip() for i in main_table.find_all('th')]
header
df = pd.DataFrame(columns=header)
df
```

```python
main_rows = main_table.find_all('tr')
main_rows = main_rows[1:]
individual_rows = []
for i in main_rows:
    individual_rows = [j.text.strip() for j in i.find_all('td')]
    length = len(df)
    df.loc[length] = individual_rows
```
