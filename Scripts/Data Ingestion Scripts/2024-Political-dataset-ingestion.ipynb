```python
df['State'] = df['State'].apply(editing_state_name)
```

```python
def editing_state_name(x):
    states_to_keep = [
        "Andaman and Nicobar Islands",
        "Arunachal Pradesh",
        "Assam",
        "Chandigarh",
        "Chhattisgarh",
        "Dadra and Nagar Haveli and Daman and Diu"
    ]
    if pd.isna(x) or x in states_to_keep :
        return x
    else: 
        return(x[:-4])
```
